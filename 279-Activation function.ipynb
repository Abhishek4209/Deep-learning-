### **Activation function:-**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the linear combination
w=[1,2] #Weights
b=3 #Bias  
```

```python
# create data
x=np.linspace(-5,5,100)
y=w[0]*x+w[1]*x+b 
```

```python
# plot linear combination
plt.plot(x,y,label="linear combination")
```

### plot sigmoid activation function

```python
# plot sigmoid activation function
sigmoid=lambda x:1/(1+np.exp(x))
y_sigmoid=sigmoid(y)
plt.plot(x,y_sigmoid,label="sigmoid Activation")
plt.legend()
plt.show()
```

### Plot ReLU activation function:-

```python
relu=lambda x: np.maximum(x,0)
y_relu= relu(y)
```

```python
plt.plot(x,y_relu,label="Relu Activation")
plt.legend()
plt.show()
```
